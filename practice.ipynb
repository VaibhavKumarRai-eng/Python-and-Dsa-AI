```python
a = 10  # 0b1010
b = 6   # 0b0110

print("a       =", a)
print("b       =", b)
print("a & b   =", a & b)   # bitwise AND
print("a | b   =", a | b)   # bitwise OR
print("a ^ b   =", a ^ b)   # bitwise XOR
print("~a      =", ~a)      # bitwise NOT
print("a << 1  =", a << 1)  # left shift
print("a >> 1  =", a >> 1)  # right shift
print("b << 2  =", b << 2)
print("b >> 1  =", b >> 1)
```

```python
def add(a, b):
    return a + b

result = add(5, 3)
print("The sum of 5 and 3 is:", result)
```

```python
def divide(a, b):
    return a / b

result = divide(10, 2)
print("The quotient of 10 and 2 is:", result)
```

```python
def multiply(a, b):
    return a * b

result = multiply(5, 3)
print("The product of 5 and 3 is:", result)
```

```python
def modulo(a, b):
    return a % b

result = modulo(10, 3)
print("The remainder of 10 and 3 is:", result)
```
